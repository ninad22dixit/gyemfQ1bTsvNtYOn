4. Switch between different strategies (e.g., day trading, swing trading, value investing)

````python
# Import necessary libraries

import sys
import os
import json
import pandas as pd
import numpy as np
from datetime import datetime, timedelta, timezone
from pathlib import Path
import warnings
import requests
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
from typing import Any, Optional, Dict, List, Tuple
````

````python
# Get historical BTC data from Coinbase API

def get_btc_historical_candles(product_id = "BTC-USD", granularity = 86400, days_back = 300):
    
    end_time = datetime.now(timezone.utc)
    start_time = end_time - timedelta(days=days_back)

    url = f"https://api.exchange.coinbase.com/products/{product_id}/candles"
    params = {"start": start_time.isoformat().replace("+00:00", "Z"),
                "end": end_time.isoformat().replace("+00:00", "Z"),
                "granularity": granularity}

    response = requests.get(url, params = params, timeout = 20)
    response.raise_for_status()
    rows = response.json()

    candles = []
    for row in rows:
        candles.append({ "time": datetime.fromtimestamp(row[0], tz=timezone.utc),
                        "low": float(row[1]), "high": float(row[2]), "open": float(row[3]),
                        "close": float(row[4]), "volume":float(row[5])})

    candles.sort(key = lambda x: x["time"])
    candles_df = pd.DataFrame(candles)

    return candles_df
````

````python
# Calculate ATR

def calculate_atr(candles_df, window = 14):

    df = candles_df.copy()
    df["prev_close"] = df["close"].shift(1)

    df["tr1"] = df["high"] - df["low"]
    df["tr2"] = (df["high"] - df["prev_close"]).abs()
    df["tr3"] = (df["low"] - df["prev_close"]).abs()

    df["true_range"] = df[["tr1", "tr2", "tr3"]].max(axis=1)
    df["ATR"] = df["true_range"].rolling(window).mean()

    return df
````

````python
# Plot BTC closing price and ATR

def plot_price_and_atr(df):

    # Ensure datetime format
    df["time"] = pd.to_datetime(df["time"])

    # Create figure
    plt.figure(figsize=(12, 8))

    # Plot BTC Price
    plt.subplot(2, 1, 1)
    plt.plot(df["time"], df["close"])
    plt.title("Bitcoin Price (Close)")
    plt.xlabel("Time")
    plt.ylabel("Price (USD)")

    # Plot ATR
    plt.subplot(2, 1, 2)
    plt.plot(df["time"], df["ATR"])
    plt.title("Average True Range (ATR)")
    plt.xlabel("Time")
    plt.ylabel("ATR (USD)")
    
    plt.tight_layout()
    plt.show()
````

````python
# Compute indicators that will prompt strategy change

def compute_indicators(df, atr_col="ATR"):

    out = df.copy().sort_values("time").reset_index(drop=True)

    required_cols = {"time", "open", "high", "low", "close", atr_col}
    missing = required_cols - set(out.columns)
    if missing:
        raise ValueError(f"Missing required columns: {sorted(missing)}")

    out["ema_20"] = out["close"].ewm(span=20, adjust=False).mean()
    out["ema_50"] = out["close"].ewm(span=50, adjust=False).mean()
    out["ema_200"] = out["close"].ewm(span=200, adjust=False).mean()

    out["atr_pct"] = out[atr_col] / out["close"]

    out["rolling_peak"] = out["close"].cummax()
    out["drawdown_from_peak"] = (out["rolling_peak"] - out["close"]) / out["rolling_peak"]

    return out
````

````python
# Select which strategy to implement - value investing (DCA) or swing trading

def detect_regime(row, drawdown_threshold=0.15):
    
    close = row["close"]
    ema_20 = row["ema_20"]
    ema_50 = row["ema_50"]
    ema_200 = row["ema_200"]
    drawdown = row["drawdown_from_peak"]

    if pd.isna(ema_20) or pd.isna(ema_50) or pd.isna(ema_200):
        return "value_investing"

    if (close < ema_200) or (drawdown >= drawdown_threshold):
        return "value_investing"

    if (close > ema_50) and (ema_20 > ema_50):
        return "swing_trading"

    return "value_investing"
````

````python
# Build profit taking targets based on ATR

def build_profit_targets(entry_price, atr_at_entry, atr_levels, sell_fractions):
    
    if len(atr_levels) != len(sell_fractions):
        raise ValueError("atr_levels and sell_fractions must have same length.")

    if abs(sum(sell_fractions) - 1.0) > 1e-9:
        raise ValueError("sell_fractions must sum to 1.0.")

    return [
        {
            "atr_level": atr_level,
            "target_price": entry_price + (atr_level * atr_at_entry),
            "sell_fraction": sell_fraction,
            "triggered": False,
            "btc_sold": 0.0,
        }
        for atr_level, sell_fraction in zip(atr_levels, sell_fractions)
    ]
````

````python
# Initialize portfolio

def initialize_portfolio(df, total_budget, dca_fraction, swing_fraction):
    remaining_cash = float(total_budget)

    portfolio = {
        "remaining_cash": remaining_cash,
        "dca_budget_limit": total_budget * dca_fraction,
        "swing_budget_limit": total_budget * swing_fraction,
        "dca_capital_deployed": 0.0,
        "swing_capital_deployed": 0.0,
        "dca_reference_price": float(df.loc[0, "close"]),
        "open_positions": [],
        "buy_events": [],
        "sell_events": [],
        "position_counter": 0
    }

    return portfolio
````

````python
#Set exit thresholds

def process_exits(row, portfolio):
    current_time = row["time"]
    candle_high = float(row["high"])
    candle_low = float(row["low"])
    regime = row["regime"]

    remaining_cash = portfolio["remaining_cash"]

    for pos in portfolio["open_positions"]:
        if pos["btc_remaining"] <= 0:
            continue

        stop_hit = candle_low <= pos["stop_loss"]

        hit_targets = [
            t for t in pos["targets"]
            if (not t["triggered"]) and candle_high >= t["target_price"]
        ]

        # Stop-loss strategy
        if stop_hit:
            btc_to_sell = pos["btc_remaining"]
            exit_price = pos["stop_loss"]

            usd_received = btc_to_sell * exit_price
            pnl = usd_received - (btc_to_sell * pos["entry_price"])

            pos["btc_remaining"] = 0.0
            remaining_cash += usd_received

            if pos["strategy"] == "value_investing":
                portfolio["dca_capital_deployed"] -= pos["usd_invested"]
            else:
                portfolio["swing_capital_deployed"] -= pos["usd_invested"]

            portfolio["sell_events"].append({
                "time": current_time,
                "strategy": pos["strategy"],
                "type": "stop_loss",
                "position_id": pos["position_id"],
                "entry_price": pos["entry_price"],
                "exit_price": exit_price,
                "btc_sold": btc_to_sell,
                "usd_received": usd_received,
                "pnl": pnl,
                "regime_at_exit": regime
            })
            continue

        # Profit taking strategy
        for target in sorted(hit_targets, key=lambda x: x["target_price"]):
            if pos["btc_remaining"] <= 0:
                break

            btc_to_sell = pos["initial_btc"] * target["sell_fraction"]
            btc_to_sell = min(btc_to_sell, pos["btc_remaining"])

            exit_price = target["target_price"]
            usd_received = btc_to_sell * exit_price
            pnl = usd_received - (btc_to_sell * pos["entry_price"])

            pos["btc_remaining"] -= btc_to_sell
            remaining_cash += usd_received

            target["triggered"] = True

            portfolio["sell_events"].append({
                "time": current_time,
                "strategy": pos["strategy"],
                "type": "take_profit",
                "position_id": pos["position_id"],
                "entry_price": pos["entry_price"],
                "exit_price": exit_price,
                "btc_sold": btc_to_sell,
                "usd_received": usd_received,
                "pnl": pnl,
                "regime_at_exit": regime
            })

    portfolio["remaining_cash"] = remaining_cash
````

````python
# DCA strategy

def process_dca_entry(row, portfolio, params):
    if row["regime"] != "value_investing":
        return

    price = row["close"]
    atr = row["ATR"]

    ref_price = portfolio["dca_reference_price"]
    drawdown = (ref_price - price) / ref_price

    if price > ref_price:
        portfolio["dca_reference_price"] = price

    if drawdown < params["drop_trigger"]:
        return

    if portfolio["remaining_cash"] < params["buy_amount"]:
        return

    if (portfolio["dca_capital_deployed"] + params["buy_amount"]) > portfolio["dca_budget_limit"]:
        return

    # Execute buy
    portfolio["position_counter"] += 1

    btc = params["buy_amount"] / price
    stop_loss = price - params["atr_mult"] * atr

    targets = build_profit_targets(
        price,
        atr,
        params["tp_levels"],
        params["tp_fracs"]
    )

    pos = {
        "position_id": portfolio["position_counter"],
        "strategy": "value_investing",
        "buy_time": row["time"],
        "entry_price": price,
        "atr_at_entry": atr,
        "stop_loss": stop_loss,
        "initial_btc": btc,
        "btc_remaining": btc,
        "usd_invested": params["buy_amount"],
        "targets": targets
    }

    portfolio["open_positions"].append(pos)
    portfolio["remaining_cash"] -= params["buy_amount"]
    portfolio["dca_capital_deployed"] += params["buy_amount"]

    portfolio["buy_events"].append({
        "time": row["time"],
        "strategy": "value_investing",
        "entry_price": price,
        "usd_spent": params["buy_amount"],
        "btc_bought": btc
    })

    portfolio["dca_reference_price"] = price
````

````python
# Swing trading strategy

def process_swing_entry(row, portfolio, params):
    if row["regime"] != "swing_trading":
        return

    price = row["close"]
    atr = row["ATR"]

    open_swings = [
        p for p in portfolio["open_positions"]
        if p["strategy"] == "swing_trading" and p["btc_remaining"] > 0
    ]

    if len(open_swings) > 0:
        return

    if portfolio["remaining_cash"] < params["buy_amount"]:
        return

    if (portfolio["swing_capital_deployed"] + params["buy_amount"]) > portfolio["swing_budget_limit"]:
        return

    if not (price > row["ema_20"] > row["ema_50"]):
        return

    # Execute buy
    portfolio["position_counter"] += 1

    btc = params["buy_amount"] / price
    stop_loss = price - params["atr_mult"] * atr

    targets = build_profit_targets(
        price,
        atr,
        params["tp_levels"],
        params["tp_fracs"]
    )

    pos = {
        "position_id": portfolio["position_counter"],
        "strategy": "swing_trading",
        "buy_time": row["time"],
        "entry_price": price,
        "atr_at_entry": atr,
        "stop_loss": stop_loss,
        "initial_btc": btc,
        "btc_remaining": btc,
        "usd_invested": params["buy_amount"],
        "targets": targets
    }

    portfolio["open_positions"].append(pos)
    portfolio["remaining_cash"] -= params["buy_amount"]
    portfolio["swing_capital_deployed"] += params["buy_amount"]

    portfolio["buy_events"].append({
        "time": row["time"],
        "strategy": "swing_trading",
        "entry_price": price,
        "usd_spent": params["buy_amount"],
        "btc_bought": btc
    })
````

````python
# Run the strategy

def run_strategy_loop(df, total_budget, dca_params, swing_params):
    
    # Compute indicators
    df = compute_indicators(df)

   # Initialize portfolio 
    portfolio = initialize_portfolio(
        df,
        total_budget,
        dca_fraction=0.75,
        swing_fraction=0.25
    )    
    
    # Assign regimes
    df["regime"] = df.apply(detect_regime, axis=1)

    for _, row in df.iterrows():
        if pd.isna(row["ATR"]) or row["ATR"] <= 0:
            continue

        # Check for exits first
        process_exits(row, portfolio)

        # Check for entries
        process_dca_entry(row, portfolio, dca_params)
        process_swing_entry(row, portfolio, swing_params)

    return portfolio, df
````

````python
# Build summary

def build_summary(portfolio, df):
    last_price = float(df.iloc[-1]["close"])

    open_positions = [
        p for p in portfolio["open_positions"]
        if p["btc_remaining"] > 0
    ]

    open_rows = []
    for pos in open_positions:
        market_value = pos["btc_remaining"] * last_price
        unrealized = market_value - pos["btc_remaining"] * pos["entry_price"]

        open_rows.append({
            "position_id": pos["position_id"],
            "strategy": pos["strategy"],
            "market_value": market_value,
            "unrealized_pnl": unrealized
        })

    open_df = pd.DataFrame(open_rows)
    sells_df = pd.DataFrame(portfolio["sell_events"])
    buys_df = pd.DataFrame(portfolio["buy_events"])

    summary = {
        "remaining_cash": portfolio["remaining_cash"],
        "total_equity": portfolio["remaining_cash"] if open_df.empty else (portfolio["remaining_cash"] + open_df["market_value"].sum()),
        "realized_pnl": sells_df["pnl"].sum() if not sells_df.empty else 0,
        "unrealized_pnl": open_df["unrealized_pnl"].sum() if not open_df.empty else 0,
        "num_trades": len(sells_df)
    }

    return buys_df, sells_df, open_df, summary
````

````python
# Execute strategy

if __name__ == "__main__":
    
    # Fetch BTC data from Coinbase API
    candles_df = get_btc_historical_candles(product_id = "BTC-USD", granularity = 86400, 
                                            days_back = 300)
    print("\nImported BTC data")
    display(candles_df.head())

    # Calculate ATR over 2 weeks
    candles_df = calculate_atr(candles_df, window = 14)
    
    print("\nBTC ATR values")
    display(candles_df.iloc[14:19])

    # Plot BTC price and ATR
    plot_price_and_atr(candles_df)

    # Set value investing and swing trading parameters
    dca_params = {"buy_amount": 500.0, "drop_trigger": 0.03, "atr_mult": 2.0,
                  "tp_levels": (1,2,3), "tp_fracs": (0.25, 0.25, 0.50)}

    swing_params = {"buy_amount": 500.0, "atr_mult": 2.0, 
                    "tp_levels": (1, 2, 3),"tp_fracs": (0.25, 0.25, 0.50)}

    portfolio, df_out = run_strategy_loop(candles_df, total_budget=10000.0, dca_params=dca_params,
                                          swing_params=swing_params)

    buys_df, sells_df, open_df, summary = build_summary(portfolio, df_out)

    # test_accounting_identity(summary, starting_budget=10000.0)
    
    print("SUMMARY")
    for k, v in summary.items():
        print(f"{k}: {v}")

    print("\nBUYS")
    display(buys_df.head())

    print("\nSELLS")
    display(sells_df.head())

    print("\nOPEN POSITIONS")
    display(open_df.head())
````


## LLM-enhanced Bitcoin signal generation

This section upgrades the notebook with:

1. **Standard technical indicators**: RSI, MACD, SMA, EMA  
2. **Bitcoin-specific on-chain features**: average block size, hash rate, mempool size  
3. **Custom engineered features**: volatility, momentum, MA distance, volume z-score, regime state  
4. **An open-source local LLM layer** using **LLaMA via Ollama** (for example `llama3.2:3b`)  
5. **LLM-guided buy / hold / sell signals** that can adapt feature selection based on context

### Local setup
Install Ollama and pull an open model locally, for example:

```bash
ollama pull llama3.2:3b
```

Then make sure Ollama is running locally on port `11434`.

### Design
The LLM does **not** replace deterministic indicators.  
Instead, it acts as a **contextual meta-decision layer** that:

- reads technical + on-chain + custom features
- selects the most relevant features for the current regime
- returns a structured JSON decision
- suggests action: `buy`, `hold`, or `sell`
- optionally adapts risk sizing and confidence thresholds


````python

# Additional imports for the LLM-enhanced workflow
import ast
from typing import Dict, List, Optional, Tuple
````

````python

# -------------------------------
# Free Bitcoin-specific data APIs
# -------------------------------

def _safe_request_json(url, params=None, timeout=20):
    try:
        resp = requests.get(url, params=params, timeout=timeout)
        resp.raise_for_status()
        return resp.json()
    except Exception as e:
        print(f"Warning: failed to fetch {url}: {e}")
        return None


def get_blockchain_chart_series(chart_name: str,
                                timespan: str = "300days",
                                rolling_average: str = "8hours") -> pd.DataFrame:
    """
    Free Blockchain.com chart endpoint.
    Example chart names:
      - hash-rate
      - avg-block-size
      - mempool-size
    """
    url = f"https://api.blockchain.info/charts/{chart_name}"
    params = {"timespan": timespan, "rollingAverage": rolling_average, "format": "json"}
    payload = _safe_request_json(url, params=params)

    if not payload or "values" not in payload:
        return pd.DataFrame(columns=["time", chart_name])

    rows = []
    for item in payload["values"]:
        rows.append({
            "time": datetime.fromtimestamp(item["x"], tz=timezone.utc).replace(hour=0, minute=0, second=0, microsecond=0),
            chart_name: float(item["y"])
        })

    out = pd.DataFrame(rows)
    out = out.groupby("time", as_index=False).mean()
    return out


def get_btc_onchain_features(days_back: int = 300) -> pd.DataFrame:
    """
    Returns daily BTC on-chain / network features merged by UTC date.
    """
    feature_map = {
        "hash-rate": "hash_rate",
        "avg-block-size": "avg_block_size",
        "mempool-size": "mempool_size"
    }

    merged = None
    for remote_name, local_name in feature_map.items():
        df_feat = get_blockchain_chart_series(remote_name, timespan=f"{days_back}days")
        if df_feat.empty:
            continue
        df_feat = df_feat.rename(columns={remote_name: local_name})
        merged = df_feat if merged is None else merged.merge(df_feat, on="time", how="outer")

    if merged is None:
        merged = pd.DataFrame(columns=["time", "hash_rate", "avg_block_size", "mempool_size"])

    merged = merged.sort_values("time").reset_index(drop=True)
    return merged
````

````python

# ---------------------------------
# Rich technical + custom features
# ---------------------------------

def compute_rsi(close: pd.Series, window: int = 14) -> pd.Series:
    delta = close.diff()
    gain = delta.clip(lower=0)
    loss = (-delta).clip(lower=0)

    avg_gain = gain.ewm(alpha=1/window, adjust=False, min_periods=window).mean()
    avg_loss = loss.ewm(alpha=1/window, adjust=False, min_periods=window).mean()

    rs = avg_gain / avg_loss.replace(0, np.nan)
    rsi = 100 - (100 / (1 + rs))
    return rsi.fillna(50)


def compute_macd(close: pd.Series,
                 fast: int = 12,
                 slow: int = 26,
                 signal: int = 9) -> Tuple[pd.Series, pd.Series, pd.Series]:
    ema_fast = close.ewm(span=fast, adjust=False).mean()
    ema_slow = close.ewm(span=slow, adjust=False).mean()
    macd_line = ema_fast - ema_slow
    signal_line = macd_line.ewm(span=signal, adjust=False).mean()
    hist = macd_line - signal_line
    return macd_line, signal_line, hist


def add_llm_features(df: pd.DataFrame, atr_col: str = "ATR") -> pd.DataFrame:
    out = df.copy().sort_values("time").reset_index(drop=True)

    # Price / trend indicators
    out["sma_20"] = out["close"].rolling(20).mean()
    out["sma_50"] = out["close"].rolling(50).mean()
    out["sma_200"] = out["close"].rolling(200).mean()

    out["ema_12"] = out["close"].ewm(span=12, adjust=False).mean()
    out["ema_20"] = out["close"].ewm(span=20, adjust=False).mean()
    out["ema_26"] = out["close"].ewm(span=26, adjust=False).mean()
    out["ema_50"] = out["close"].ewm(span=50, adjust=False).mean()
    out["ema_200"] = out["close"].ewm(span=200, adjust=False).mean()

    # RSI / MACD
    out["rsi_14"] = compute_rsi(out["close"], window=14)
    out["macd"], out["macd_signal"], out["macd_hist"] = compute_macd(out["close"])

    # ATR context
    out["atr_pct"] = out[atr_col] / out["close"]

    # Return and momentum features
    out["ret_1d"] = out["close"].pct_change(1)
    out["ret_3d"] = out["close"].pct_change(3)
    out["ret_7d"] = out["close"].pct_change(7)
    out["ret_14d"] = out["close"].pct_change(14)

    out["mom_3d"] = out["close"] / out["close"].shift(3) - 1
    out["mom_7d"] = out["close"] / out["close"].shift(7) - 1
    out["mom_14d"] = out["close"] / out["close"].shift(14) - 1

    # Volatility / volume
    out["realized_vol_14"] = out["ret_1d"].rolling(14).std() * np.sqrt(14)
    out["volume_sma_20"] = out["volume"].rolling(20).mean()
    out["volume_zscore_20"] = (
        (out["volume"] - out["volume"].rolling(20).mean()) /
        out["volume"].rolling(20).std()
    )

    # Distance to moving averages
    out["dist_to_sma20"] = (out["close"] / out["sma_20"]) - 1
    out["dist_to_sma50"] = (out["close"] / out["sma_50"]) - 1
    out["dist_to_ema20"] = (out["close"] / out["ema_20"]) - 1
    out["dist_to_ema50"] = (out["close"] / out["ema_50"]) - 1
    out["trend_spread_20_50"] = out["ema_20"] - out["ema_50"]
    out["trend_spread_50_200"] = out["ema_50"] - out["ema_200"]

    # Drawdown / breakout context
    out["rolling_peak"] = out["close"].cummax()
    out["drawdown_from_peak"] = (out["rolling_peak"] - out["close"]) / out["rolling_peak"]
    out["rolling_low_30"] = out["close"].rolling(30).min()
    out["rolling_high_30"] = out["close"].rolling(30).max()
    out["breakout_pct_30"] = (out["close"] - out["rolling_high_30"]) / out["rolling_high_30"]
    out["rebound_from_low_30"] = (out["close"] - out["rolling_low_30"]) / out["rolling_low_30"]

    # Bitcoin-specific network feature transforms
    for col in ["hash_rate", "avg_block_size", "mempool_size"]:
        if col not in out.columns:
            out[col] = np.nan

        out[f"{col}_chg_7d"] = out[col].pct_change(7)
        out[f"{col}_zscore_30"] = (
            (out[col] - out[col].rolling(30).mean()) /
            out[col].rolling(30).std()
        )

    # Composite custom features
    out["trend_score"] = (
        0.35 * (out["ema_20"] > out["ema_50"]).astype(float) +
        0.35 * (out["ema_50"] > out["ema_200"]).astype(float) +
        0.30 * (out["macd_hist"] > 0).astype(float)
    )

    out["mean_reversion_score"] = (
        (out["rsi_14"] < 35).astype(float) +
        (out["drawdown_from_peak"] > 0.10).astype(float) +
        (out["dist_to_sma20"] < -0.03).astype(float)
    )

    out["onchain_strength_score"] = (
        (out["hash_rate_chg_7d"] > 0).astype(float) +
        (out["avg_block_size_chg_7d"] > 0).astype(float) +
        (out["mempool_size_zscore_30"] > 0).astype(float)
    ) / 3.0

    return out


def prepare_market_context(days_back: int = 300) -> pd.DataFrame:
    candles_df = get_btc_historical_candles(product_id="BTC-USD", granularity=86400, days_back=days_back)
    candles_df = calculate_atr(candles_df, window=14)

    onchain_df = get_btc_onchain_features(days_back=days_back)

    merged = candles_df.merge(onchain_df, on="time", how="left")
    merged = merged.sort_values("time").reset_index(drop=True)

    # Forward fill slow-moving network metrics
    for col in ["hash_rate", "avg_block_size", "mempool_size"]:
        if col in merged.columns:
            merged[col] = merged[col].ffill()

    merged = add_llm_features(merged)
    merged["regime"] = merged.apply(detect_regime, axis=1)
    return merged
````


## Hybrid LLM + Quantitative Trading System

This version uses the local LLaMA/Ollama model only for slower strategic decisions:

1. **Regime switching**: choose `value_investing`, `swing_trading`, or `defensive` for the next review window.
2. **Feature selection**: identify which indicators/features matter most in the current market context.
3. **Risk adjustment**: adapt buy size, stop-loss width, take-profit distance, and whether new entries are allowed.

Per-candle execution remains deterministic and rule-based. This avoids slow, noisy LLM calls on every candle.


````python

# ---------------------------------
# Local LLaMA / Ollama helper
# ---------------------------------

def query_ollama(prompt: str,
                 model: str = "llama3.2:3b",
                 url: str = "http://localhost:11434/api/generate",
                 timeout: int = 180) -> Optional[Dict[str, Any]]:
    """
    Calls a local Ollama model. Returns parsed JSON or None.
    If Ollama is unavailable, the hybrid strategy automatically falls back to heuristics.
    """
    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False,
        "options": {
            "temperature": 0.1,
            "num_predict": 250
        }
    }

    try:
        resp = requests.post(url, json=payload, timeout=timeout)
        resp.raise_for_status()
        data = resp.json()
        raw = data.get("response", "").strip()
        if not raw:
            return None
        return json.loads(raw)
    except Exception as e:
        print(f"Warning: local LLM call failed; using heuristic fallback. Error: {e}")
        return None
````

````python

# ---------------------------------
# Hybrid LLM + Quantitative strategy layer
# ---------------------------------

HYBRID_FEATURES = [
    "close", "ATR", "atr_pct", "rsi_14",
    "macd", "macd_signal", "macd_hist",
    "sma_20", "sma_50", "sma_200",
    "ema_20", "ema_50", "ema_200",
    "ret_3d", "ret_7d", "ret_14d",
    "mom_7d", "mom_14d",
    "realized_vol_14", "volume_zscore_20",
    "drawdown_from_peak", "dist_to_sma20", "dist_to_sma50",
    "trend_score", "mean_reversion_score",
    "hash_rate_chg_7d", "avg_block_size_chg_7d", "mempool_size_zscore_30",
    "onchain_strength_score",
]


def safe_float(value, default: float) -> float:
    """Safely convert values returned by the LLM into finite floats."""
    try:
        if value is None:
            return default
        value = float(value)
        if np.isnan(value) or np.isinf(value):
            return default
        return value
    except (TypeError, ValueError):
        return default


def safe_bool(value, default: bool = False) -> bool:
    """Safely parse booleans returned by the LLM."""
    if isinstance(value, bool):
        return value
    if value is None:
        return default
    if isinstance(value, str):
        return value.strip().lower() in {"true", "yes", "1", "y"}
    return bool(value)


def clamp(value: float, low: float, high: float) -> float:
    return max(low, min(high, value))


def summarize_recent_market_window(df: pd.DataFrame,
                                   end_idx: int,
                                   lookback: int = 30,
                                   feature_cols: Optional[List[str]] = None) -> Dict[str, Any]:
    """
    Compress the last N candles into a compact JSON-safe market summary.
    The LLM sees this summary instead of every candle.
    """
    feature_cols = feature_cols or HYBRID_FEATURES
    start_idx = max(0, end_idx - lookback + 1)
    window = df.iloc[start_idx:end_idx + 1].copy()
    latest = window.iloc[-1]

    latest_snapshot = {}
    for col in feature_cols:
        if col in latest.index:
            val = latest[col]
            if pd.isna(val):
                latest_snapshot[col] = None
            elif isinstance(val, (np.integer, int)):
                latest_snapshot[col] = int(val)
            elif isinstance(val, (np.floating, float)):
                latest_snapshot[col] = round(float(val), 6)
            else:
                latest_snapshot[col] = str(val)

    def last_valid(col: str, default=None):
        if col not in window.columns:
            return default
        s = window[col].dropna()
        if len(s) == 0:
            return default
        return s.iloc[-1]

    summary = {
        "time": str(latest.get("time", "")),
        "lookback_days": int(len(window)),
        "latest": latest_snapshot,
        "window_stats": {
            "price_return": safe_float(window["close"].iloc[-1] / window["close"].iloc[0] - 1, 0.0) if len(window) > 1 else 0.0,
            "price_max_drawdown": safe_float(((window["close"].cummax() - window["close"]) / window["close"].cummax()).max(), 0.0),
            "avg_atr_pct": safe_float(window["atr_pct"].mean(), 0.0) if "atr_pct" in window.columns else 0.0,
            "avg_rsi_14": safe_float(window["rsi_14"].mean(), 50.0) if "rsi_14" in window.columns else 50.0,
            "latest_macd_hist": safe_float(last_valid("macd_hist", 0.0), 0.0),
            "latest_trend_score": safe_float(last_valid("trend_score", 0.0), 0.0),
            "latest_mean_reversion_score": safe_float(last_valid("mean_reversion_score", 0.0), 0.0),
            "latest_onchain_strength_score": safe_float(last_valid("onchain_strength_score", 0.0), 0.0),
        },
        "baseline_rule_regime": str(latest.get("regime", "value_investing")),
    }
    return summary


def build_hybrid_llm_prompt(market_summary: Dict[str, Any]) -> str:
    schema = {
        "target_regime": "value_investing | swing_trading | defensive",
        "confidence": "float between 0 and 1",
        "selected_features": ["list", "of", "most", "useful", "features"],
        "reasoning": "brief explanation",
        "risk_adjustment": {
            "dca_buy_multiplier": "numeric, 0.5 to 1.5",
            "swing_buy_multiplier": "numeric, 0.5 to 1.5",
            "stop_atr_multiplier": "numeric, 1.25 to 3.5",
            "take_profit_multiplier": "numeric, 0.75 to 1.5",
            "max_position_multiplier": "numeric, 0.25 to 1.0",
            "allow_new_entries": "true or false",
            "force_reduce_exposure": "true or false"
        }
    }

    prompt = f"""
You are a local open-source LLaMA crypto strategy supervisor.

You do NOT execute trades directly. Your job is only to set the strategy context for the next review window.
The deterministic trading rules will execute every candle.

Choose exactly one target_regime:
- value_investing: BTC is weak, oversold, or in drawdown; prefer DCA-style accumulation only after rule-based pullback triggers.
- swing_trading: BTC is trending upward; allow rule-based trend-following entries.
- defensive: risk is elevated; reduce new entries and tighten risk.

Return ONLY valid JSON. Do not use markdown.
All risk_adjustment fields must be numeric or boolean. Never return null.

Market summary:
{json.dumps(market_summary, indent=2)}

Return schema:
{json.dumps(schema, indent=2)}
""".strip()
    return prompt


def heuristic_hybrid_policy(market_summary: Dict[str, Any]) -> Dict[str, Any]:
    """Deterministic fallback when Ollama is unavailable or returns invalid JSON."""
    latest = market_summary.get("latest", {})
    stats = market_summary.get("window_stats", {})

    rsi = safe_float(latest.get("rsi_14"), 50.0)
    macd_hist = safe_float(latest.get("macd_hist"), 0.0)
    trend_score = safe_float(latest.get("trend_score"), 0.0)
    mean_rev = safe_float(latest.get("mean_reversion_score"), 0.0)
    drawdown = safe_float(latest.get("drawdown_from_peak"), 0.0)
    atr_pct = safe_float(latest.get("atr_pct"), 0.03)
    onchain = safe_float(latest.get("onchain_strength_score"), 0.0)
    window_return = safe_float(stats.get("price_return"), 0.0)

    selected = ["rsi_14", "macd_hist", "trend_score", "mean_reversion_score", "drawdown_from_peak", "atr_pct", "onchain_strength_score"]

    if atr_pct > 0.07 or (macd_hist < 0 and window_return < -0.08 and trend_score < 0.35):
        regime = "defensive"
        confidence = 0.72
        risk = {
            "dca_buy_multiplier": 0.5,
            "swing_buy_multiplier": 0.5,
            "stop_atr_multiplier": 1.5,
            "take_profit_multiplier": 0.85,
            "max_position_multiplier": 0.5,
            "allow_new_entries": False,
            "force_reduce_exposure": True,
        }
    elif trend_score >= 0.70 and macd_hist > 0 and rsi < 75:
        regime = "swing_trading"
        confidence = 0.68 + 0.12 * min(1.0, trend_score)
        risk = {
            "dca_buy_multiplier": 0.75,
            "swing_buy_multiplier": 1.25 if onchain >= 0.5 else 1.0,
            "stop_atr_multiplier": 2.25,
            "take_profit_multiplier": 1.15,
            "max_position_multiplier": 1.0,
            "allow_new_entries": True,
            "force_reduce_exposure": False,
        }
    elif rsi < 40 or drawdown > 0.10 or mean_rev >= 1.0:
        regime = "value_investing"
        confidence = 0.66
        risk = {
            "dca_buy_multiplier": 1.25 if rsi < 35 and onchain >= 0.33 else 1.0,
            "swing_buy_multiplier": 0.5,
            "stop_atr_multiplier": 2.75,
            "take_profit_multiplier": 1.0,
            "max_position_multiplier": 0.75,
            "allow_new_entries": True,
            "force_reduce_exposure": False,
        }
    else:
        regime = market_summary.get("baseline_rule_regime", "value_investing")
        confidence = 0.55
        risk = {
            "dca_buy_multiplier": 1.0,
            "swing_buy_multiplier": 1.0,
            "stop_atr_multiplier": 2.0,
            "take_profit_multiplier": 1.0,
            "max_position_multiplier": 0.75,
            "allow_new_entries": True,
            "force_reduce_exposure": False,
        }

    return {
        "target_regime": regime,
        "confidence": clamp(confidence, 0.0, 1.0),
        "selected_features": selected,
        "reasoning": "Heuristic hybrid policy based on trend, mean reversion, volatility, and on-chain strength.",
        "risk_adjustment": risk,
    }


def get_hybrid_llm_policy(market_summary: Dict[str, Any],
                          model: str = "llama3.2:3b",
                          min_confidence: float = 0.55) -> Dict[str, Any]:
    prompt = build_hybrid_llm_prompt(market_summary)
    llm_out = query_ollama(prompt, model=model)

    if llm_out is None:
        llm_out = heuristic_hybrid_policy(market_summary)

    regime = str(llm_out.get("target_regime", "value_investing")).lower().strip()
    if regime not in {"value_investing", "swing_trading", "defensive"}:
        regime = "value_investing"

    confidence = clamp(safe_float(llm_out.get("confidence"), 0.5), 0.0, 1.0)
    if confidence < min_confidence:
        # Low-confidence LLM views are allowed to influence feature selection, but not regime.
        regime = str(market_summary.get("baseline_rule_regime", "value_investing"))

    risk = llm_out.get("risk_adjustment", {}) or {}
    risk_clean = {
        "dca_buy_multiplier": clamp(safe_float(risk.get("dca_buy_multiplier"), 1.0), 0.5, 1.5),
        "swing_buy_multiplier": clamp(safe_float(risk.get("swing_buy_multiplier"), 1.0), 0.5, 1.5),
        "stop_atr_multiplier": clamp(safe_float(risk.get("stop_atr_multiplier"), 2.0), 1.25, 3.5),
        "take_profit_multiplier": clamp(safe_float(risk.get("take_profit_multiplier"), 1.0), 0.75, 1.5),
        "max_position_multiplier": clamp(safe_float(risk.get("max_position_multiplier"), 0.75), 0.25, 1.0),
        "allow_new_entries": safe_bool(risk.get("allow_new_entries"), True),
        "force_reduce_exposure": safe_bool(risk.get("force_reduce_exposure"), False),
    }

    return {
        "strategy_regime": regime,
        "hybrid_confidence": confidence,
        "hybrid_selected_features": llm_out.get("selected_features", []),
        "hybrid_reasoning": llm_out.get("reasoning", ""),
        **risk_clean,
    }


def attach_hybrid_policies(df: pd.DataFrame,
                           model: str = "llama3.2:3b",
                           review_frequency: int = 7,
                           lookback: int = 30,
                           min_confidence: float = 0.55) -> pd.DataFrame:
    """
    Calls the LLM only every `review_frequency` rows, then forward-fills that policy.
    Example: 300 candles with review_frequency=7 requires about 43 LLM calls, not 300.
    """
    out = df.copy().sort_values("time").reset_index(drop=True)

    policy_defaults = {
        "strategy_regime": out.get("regime", pd.Series(["value_investing"] * len(out))).astype(str),
        "hybrid_confidence": 0.5,
        "hybrid_selected_features": [[] for _ in range(len(out))],
        "hybrid_reasoning": "Initial default policy before first LLM review.",
        "dca_buy_multiplier": 1.0,
        "swing_buy_multiplier": 1.0,
        "stop_atr_multiplier": 2.0,
        "take_profit_multiplier": 1.0,
        "max_position_multiplier": 0.75,
        "allow_new_entries": True,
        "force_reduce_exposure": False,
        "llm_review_point": False,
    }

    for col, default in policy_defaults.items():
        out[col] = default

    last_policy = None
    for i in range(len(out)):
        enough_history = i >= max(14, min(lookback, len(out) - 1))
        should_review = enough_history and ((i % review_frequency == 0) or last_policy is None)

        if should_review:
            summary = summarize_recent_market_window(out, end_idx=i, lookback=lookback)
            last_policy = get_hybrid_llm_policy(summary, model=model, min_confidence=min_confidence)
            last_policy["llm_review_point"] = True
        elif last_policy is not None:
            last_policy = {**last_policy, "llm_review_point": False}

        if last_policy is not None:
            for key, value in last_policy.items():
                out.at[i, key] = value

    return out
````

````python

# ---------------------------------
# Rule-based per-candle execution using hybrid LLM policy
# ---------------------------------

def scale_targets_by_llm(params: Dict[str, Any], take_profit_multiplier: float) -> Tuple[Tuple[float, ...], Tuple[float, ...]]:
    tp_levels = tuple(float(x) * float(take_profit_multiplier) for x in params["tp_levels"])
    tp_fracs = tuple(float(x) for x in params["tp_fracs"])
    return tp_levels, tp_fracs


def current_strategy_exposure(portfolio: Dict[str, Any], strategy: str) -> float:
    return sum(
        float(p.get("usd_invested", 0.0))
        for p in portfolio["open_positions"]
        if p.get("strategy") == strategy and p.get("btc_remaining", 0.0) > 0
    )


def process_hybrid_risk_reduction(row: pd.Series, portfolio: Dict[str, Any], sell_fraction: float = 0.50):
    """
    Deterministic risk rule: if the LLM policy says risk is elevated, reduce exposure only when
    price/momentum confirmation is bearish. This keeps per-candle exits rule-based.
    """
    if not bool(row.get("force_reduce_exposure", False)):
        return

    bearish_confirmation = (
        float(row.get("close", np.nan)) < float(row.get("ema_20", np.inf)) or
        float(row.get("macd_hist", 0.0)) < 0 or
        str(row.get("strategy_regime", "")) == "defensive"
    )
    if not bearish_confirmation:
        return

    current_time = row["time"]
    exit_price = float(row["close"])
    remaining_cash = portfolio["remaining_cash"]

    for pos in portfolio["open_positions"]:
        if pos.get("btc_remaining", 0.0) <= 0:
            continue

        btc_to_sell = min(pos["btc_remaining"], pos["btc_remaining"] * sell_fraction)
        usd_received = btc_to_sell * exit_price
        pnl = usd_received - (btc_to_sell * pos["entry_price"])

        pos["btc_remaining"] -= btc_to_sell
        remaining_cash += usd_received

        capital_fraction_closed = btc_to_sell / max(pos["initial_btc"], 1e-12)
        capital_reduction = pos["usd_invested"] * capital_fraction_closed

        if pos["strategy"] == "value_investing":
            portfolio["dca_capital_deployed"] = max(0.0, portfolio["dca_capital_deployed"] - capital_reduction)
        else:
            portfolio["swing_capital_deployed"] = max(0.0, portfolio["swing_capital_deployed"] - capital_reduction)

        portfolio["sell_events"].append({
            "time": current_time,
            "strategy": pos["strategy"],
            "type": "hybrid_risk_reduction",
            "position_id": pos["position_id"],
            "entry_price": pos["entry_price"],
            "exit_price": exit_price,
            "btc_sold": btc_to_sell,
            "usd_received": usd_received,
            "pnl": pnl,
            "regime_at_exit": row.get("strategy_regime", row.get("regime", "unknown")),
            "hybrid_confidence": row.get("hybrid_confidence", np.nan),
            "hybrid_reasoning": row.get("hybrid_reasoning", ""),
        })

    portfolio["remaining_cash"] = remaining_cash


def process_dca_entry_hybrid(row: pd.Series, portfolio: Dict[str, Any], params: Dict[str, Any]):
    """Rule-based DCA entry, gated and sized by slower LLM policy."""
    if str(row.get("strategy_regime", row.get("regime", "value_investing"))) != "value_investing":
        return
    if not bool(row.get("allow_new_entries", True)):
        return

    price = float(row["close"])
    atr = float(row["ATR"])
    ref_price = float(portfolio["dca_reference_price"])
    drawdown = (ref_price - price) / ref_price if ref_price > 0 else 0.0

    if price > ref_price:
        portfolio["dca_reference_price"] = price

    # Deterministic DCA trigger: buy only after a configured pullback from reference price.
    if drawdown < params["drop_trigger"]:
        return

    buy_multiplier = clamp(safe_float(row.get("dca_buy_multiplier"), 1.0), 0.5, 1.5)
    buy_amount = float(params["buy_amount"]) * buy_multiplier

    # LLM adjusts max exposure, but rule-based budget checks still enforce safety.
    max_exposure = portfolio["dca_budget_limit"] * clamp(safe_float(row.get("max_position_multiplier"), 0.75), 0.25, 1.0)
    if current_strategy_exposure(portfolio, "value_investing") + buy_amount > max_exposure:
        return
    if portfolio["remaining_cash"] < buy_amount:
        return

    stop_mult = clamp(safe_float(row.get("stop_atr_multiplier"), params["atr_mult"]), 1.25, 3.5)
    tp_mult = clamp(safe_float(row.get("take_profit_multiplier"), 1.0), 0.75, 1.5)
    tp_levels, tp_fracs = scale_targets_by_llm(params, tp_mult)

    portfolio["position_counter"] += 1
    btc = buy_amount / price
    stop_loss = price - stop_mult * atr
    targets = build_profit_targets(price, atr, tp_levels, tp_fracs)

    pos = {
        "position_id": portfolio["position_counter"],
        "strategy": "value_investing",
        "buy_time": row["time"],
        "entry_price": price,
        "atr_at_entry": atr,
        "stop_loss": stop_loss,
        "initial_btc": btc,
        "btc_remaining": btc,
        "usd_invested": buy_amount,
        "targets": targets,
        "hybrid_reasoning": row.get("hybrid_reasoning", ""),
        "hybrid_selected_features": row.get("hybrid_selected_features", []),
    }

    portfolio["open_positions"].append(pos)
    portfolio["remaining_cash"] -= buy_amount
    portfolio["dca_capital_deployed"] += buy_amount
    portfolio["buy_events"].append({
        "time": row["time"],
        "strategy": "value_investing",
        "entry_price": price,
        "usd_spent": buy_amount,
        "btc_bought": btc,
        "rule_trigger": "dca_pullback",
        "strategy_regime": row.get("strategy_regime", "value_investing"),
        "hybrid_confidence": row.get("hybrid_confidence", np.nan),
        "hybrid_selected_features": row.get("hybrid_selected_features", []),
        "hybrid_reasoning": row.get("hybrid_reasoning", ""),
    })
    portfolio["dca_reference_price"] = price


def process_swing_entry_hybrid(row: pd.Series, portfolio: Dict[str, Any], params: Dict[str, Any]):
    """Rule-based swing entry, gated and sized by slower LLM policy."""
    if str(row.get("strategy_regime", row.get("regime", "value_investing"))) != "swing_trading":
        return
    if not bool(row.get("allow_new_entries", True)):
        return

    price = float(row["close"])
    atr = float(row["ATR"])

    # Deterministic swing trigger: trend-following confirmation.
    trend_ok = (
        price > float(row.get("ema_20", np.inf)) > float(row.get("ema_50", np.inf)) and
        float(row.get("macd_hist", 0.0)) > 0 and
        float(row.get("rsi_14", 50.0)) < 75
    )
    if not trend_ok:
        return

    open_swings = [
        p for p in portfolio["open_positions"]
        if p["strategy"] == "swing_trading" and p["btc_remaining"] > 0
    ]
    if len(open_swings) > 0:
        return

    buy_multiplier = clamp(safe_float(row.get("swing_buy_multiplier"), 1.0), 0.5, 1.5)
    buy_amount = float(params["buy_amount"]) * buy_multiplier

    max_exposure = portfolio["swing_budget_limit"] * clamp(safe_float(row.get("max_position_multiplier"), 0.75), 0.25, 1.0)
    if current_strategy_exposure(portfolio, "swing_trading") + buy_amount > max_exposure:
        return
    if portfolio["remaining_cash"] < buy_amount:
        return

    stop_mult = clamp(safe_float(row.get("stop_atr_multiplier"), params["atr_mult"]), 1.25, 3.5)
    tp_mult = clamp(safe_float(row.get("take_profit_multiplier"), 1.0), 0.75, 1.5)
    tp_levels, tp_fracs = scale_targets_by_llm(params, tp_mult)

    portfolio["position_counter"] += 1
    btc = buy_amount / price
    stop_loss = price - stop_mult * atr
    targets = build_profit_targets(price, atr, tp_levels, tp_fracs)

    pos = {
        "position_id": portfolio["position_counter"],
        "strategy": "swing_trading",
        "buy_time": row["time"],
        "entry_price": price,
        "atr_at_entry": atr,
        "stop_loss": stop_loss,
        "initial_btc": btc,
        "btc_remaining": btc,
        "usd_invested": buy_amount,
        "targets": targets,
        "hybrid_reasoning": row.get("hybrid_reasoning", ""),
        "hybrid_selected_features": row.get("hybrid_selected_features", []),
    }

    portfolio["open_positions"].append(pos)
    portfolio["remaining_cash"] -= buy_amount
    portfolio["swing_capital_deployed"] += buy_amount
    portfolio["buy_events"].append({
        "time": row["time"],
        "strategy": "swing_trading",
        "entry_price": price,
        "usd_spent": buy_amount,
        "btc_bought": btc,
        "rule_trigger": "trend_confirmation",
        "strategy_regime": row.get("strategy_regime", "swing_trading"),
        "hybrid_confidence": row.get("hybrid_confidence", np.nan),
        "hybrid_selected_features": row.get("hybrid_selected_features", []),
        "hybrid_reasoning": row.get("hybrid_reasoning", ""),
    })
````

````python

# ---------------------------------
# End-to-end hybrid LLM + quantitative strategy
# ---------------------------------

def run_hybrid_llm_quant_strategy(total_budget: float = 10000.0,
                                  days_back: int = 300,
                                  dca_params: Optional[Dict[str, Any]] = None,
                                  swing_params: Optional[Dict[str, Any]] = None,
                                  llm_model="llama3.2:1b",
                                  llm_review_frequency: int = 7,
                                  llm_lookback: int = 30,
                                  llm_min_confidence: float = 0.55) -> Tuple[Dict[str, Any], pd.DataFrame]:
    """
    Hybrid design:
    - LLM runs every N candles for regime switching, feature selection, and risk adjustment.
    - Deterministic rules run on every candle for entries and exits.
    """
    if dca_params is None:
        dca_params = {
            "buy_amount": 500.0,
            "drop_trigger": 0.03,
            "atr_mult": 2.0,
            "tp_levels": (1, 2, 3),
            "tp_fracs": (0.25, 0.25, 0.50),
        }

    if swing_params is None:
        swing_params = {
            "buy_amount": 500.0,
            "atr_mult": 2.0,
            "tp_levels": (1, 2, 3),
            "tp_fracs": (0.25, 0.25, 0.50),
        }

    df = prepare_market_context(days_back=days_back)
    df = attach_hybrid_policies(
        df,
        model=llm_model,
        review_frequency=llm_review_frequency,
        lookback=llm_lookback,
        min_confidence=llm_min_confidence,
    )

    portfolio = initialize_portfolio(
        df,
        total_budget=total_budget,
        dca_fraction=0.75,
        swing_fraction=0.25,
    )

    for _, row in df.iterrows():
        if pd.isna(row["ATR"]) or row["ATR"] <= 0:
            continue

        # 1. Rule-based stops and profit-taking on every candle.
        process_exits(row, portfolio)

        # 2. Rule-based risk reduction, gated by the slower LLM risk policy.
        process_hybrid_risk_reduction(row, portfolio, sell_fraction=0.50)

        # 3. Rule-based entries, gated/sized by the slower LLM strategy policy.
        process_dca_entry_hybrid(row, portfolio, dca_params)
        process_swing_entry_hybrid(row, portfolio, swing_params)

    return portfolio, df


def summarize_hybrid_policy(df: pd.DataFrame) -> pd.DataFrame:
    cols = [
        "time", "close", "regime", "strategy_regime", "llm_review_point",
        "hybrid_confidence", "allow_new_entries", "force_reduce_exposure",
        "dca_buy_multiplier", "swing_buy_multiplier", "stop_atr_multiplier",
        "take_profit_multiplier", "max_position_multiplier",
        "hybrid_selected_features", "hybrid_reasoning",
    ]
    keep = [c for c in cols if c in df.columns]
    return df[keep].copy()
````

````python

# ---------------------------------
# Execute the hybrid LLM + quantitative strategy
# ---------------------------------

if __name__ == "__main__":
    portfolio_hybrid, df_hybrid = run_hybrid_llm_quant_strategy(
        total_budget=10000.0,
        days_back=300,
        llm_model="llama3.2:1b",
        llm_review_frequency=10,   # LLM call about once per week instead of every candle
        llm_lookback=30,
        llm_min_confidence=0.55,
    )

    buys_df, sells_df, open_df, summary = build_summary(portfolio_hybrid, df_hybrid)
    policy_df = summarize_hybrid_policy(df_hybrid)

    print("HYBRID LLM + QUANT SUMMARY")
    for k, v in summary.items():
        print(f"{k}: {v}")

    print("\nLATEST HYBRID POLICIES")
    display(policy_df.tail(15))

    print("\nLLM REVIEW POINTS")
    display(policy_df[policy_df["llm_review_point"] == True].tail(10))

    print("\nBUYS")
    display(buys_df.tail(10))

    print("\nSELLS")
    display(sells_df.tail(10))

    print("\nOPEN POSITIONS")
    display(open_df.head())
````
