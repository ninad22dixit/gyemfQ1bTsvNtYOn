## LLM-enhanced Bitcoin signal generation

This notebook uses the following:

1. **Standard technical indicators**: RSI, MACD, SMA, EMA  
2. **Bitcoin-specific on-chain features**: average block size, hash rate, mempool size  
3. **Custom engineered features**: volatility, momentum, MA distance, volume z-score, regime state  
4. **An open-source local LLM layer** using **LLaMA via Ollama** (`llama3.2:1b`)  
5. **LLM-guided buy / hold / sell signals** that can adapt feature selection based on context

````python
# Import necessary libraries

import sys
import os
import json
import pandas as pd
import numpy as np
from datetime import datetime, timedelta, timezone
from pathlib import Path
import warnings
import requests
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
from typing import Any, Optional, Dict, List, Tuple
import ast
from typing import Dict, List, Optional, Tuple
````

````python
# Get historical BTC data from Coinbase API

def get_btc_historical_candles(product_id = "BTC-USD", granularity = 86400, days_back = 300):
    
    end_time = datetime.now(timezone.utc)
    start_time = end_time - timedelta(days=days_back)

    url = f"https://api.exchange.coinbase.com/products/{product_id}/candles"
    params = {"start": start_time.isoformat().replace("+00:00", "Z"),
                "end": end_time.isoformat().replace("+00:00", "Z"),
                "granularity": granularity}

    response = requests.get(url, params = params, timeout = 20)
    response.raise_for_status()
    rows = response.json()

    candles = []
    for row in rows:
        candles.append({ "time": datetime.fromtimestamp(row[0], tz=timezone.utc),
                        "low": float(row[1]), "high": float(row[2]), "open": float(row[3]),
                        "close": float(row[4]), "volume":float(row[5])})

    candles.sort(key = lambda x: x["time"])
    candles_df = pd.DataFrame(candles)

    return candles_df
````

````python
# Calculate ATR

def calculate_atr(candles_df, window = 14):

    df = candles_df.copy()
    df["prev_close"] = df["close"].shift(1)

    df["tr1"] = df["high"] - df["low"]
    df["tr2"] = (df["high"] - df["prev_close"]).abs()
    df["tr3"] = (df["low"] - df["prev_close"]).abs()

    df["true_range"] = df[["tr1", "tr2", "tr3"]].max(axis=1)
    df["ATR"] = df["true_range"].rolling(window).mean()

    return df
````

````python
# Plot BTC closing price and ATR

def plot_price_and_atr(df):

    # Ensure datetime format
    df["time"] = pd.to_datetime(df["time"])

    # Create figure
    plt.figure(figsize=(12, 8))

    # Plot BTC Price
    plt.subplot(2, 1, 1)
    plt.plot(df["time"], df["close"])
    plt.title("Bitcoin Price (Close)")
    plt.xlabel("Time")
    plt.ylabel("Price (USD)")

    # Plot ATR
    plt.subplot(2, 1, 2)
    plt.plot(df["time"], df["ATR"])
    plt.title("Average True Range (ATR)")
    plt.xlabel("Time")
    plt.ylabel("ATR (USD)")
    
    plt.tight_layout()
    plt.show()
````

````python
# Select which strategy to implement - value investing (DCA) or swing trading

def detect_regime(row, drawdown_threshold=0.15):
    
    close = row["close"]
    ema_20 = row["ema_20"]
    ema_50 = row["ema_50"]
    ema_200 = row["ema_200"]
    drawdown = row["drawdown_from_peak"]

    if pd.isna(ema_20) or pd.isna(ema_50) or pd.isna(ema_200):
        return "value_investing"

    if (close < ema_200) or (drawdown >= drawdown_threshold):
        return "value_investing"

    if (close > ema_50) and (ema_20 > ema_50):
        return "swing_trading"

    return "value_investing"
````

````python
# Import Bitcoin data

def safe_request_json(url, params = None, timeout = 20):
    try:
        resp = requests.get(url, params = params, timeout = timeout)
        resp = raise_for_status()
        return resp.json()
    except Exception as e:
        print(f"Warning: failed to fetch {url}: {e}")
        return None
````

````python
def get_blockchain_chart_series(chart_name, timespan = "300days", rolling_average = "8hours"):
    
    url = f"https://api.blockchain.info/charts/{chart_name}"
    params = {"timespan": timespan, "rollingAverage": rolling_average, "format": "json"}
    payload = safe_request_json(url, params = params)

    if not payload or "values" not in payload:
        return pd.DataFrame(columns = ["time", chart_name])

    rows = []
    for item in payload["values"]:
        rows.append({"time": datetime.fromtimestamp(item["x"], tz = timezone.utc).replace(hour = 0, minute = 0, second = 0, microsecond = 0),
                     chart_name: float(item["y"])})

    out = pd.DataFrame(rows)
    out = out.groupby("time", as_index = False).mean()
    return out
````

````python
def get_btc_onchain_features(days_back = 300):

    feature_map = {"hash-rate": "hash_rate", "avg-block-size": "avg_block_size",
                   "mempool-size": "mempool_size"}

    merged = None
    for remote_name, local_name in feature_map.items():
        df_feat = get_blockchain_chart_series(remote_name, timespan = f"{days_back}days")
        if df_feat.empty:
            continue
        df_feat = df_feat.rename(columns = {remote_name: local_name})
        merged = df_feat if merged is None else merged.merge(df_feat, on = "time", how = "outer")

    if merged is None:
        merged = pd.DataFrame(columns = ["time", "hash_rate", "avg_block_size", "mempool_size"])

    merged = merged.sort_values("time").reset_index(drop = True)
    return merged
````

````python
# Calculate technical features

def compute_rsi(close, window = 14):
    
    delta = close.diff()
    gain = delta.clip(lower = 0)
    loss = (-delta).clip(lower = 0)

    avg_gain = gain.ewm(alpha = 1/window, adjust = False, min_period = window).mean()
    avg_loss = loss.ewm(alpha = 1/window, adjust = False, min_period = window).mean()

    rs = avg_gain / avg_loss.replace(0, np.nan)
    rsi = 100 - (100 / (1 + rs))
    return rsi.fillna(50)
````

````python
def compute_macd(close, fast = 12, slow = 26, signal = 9):
    
    ema_fast = close.ewm(span = fast, adjust = False).mean()
    ema_slow = close.ewm(span = slow, adjust = False).mean()
    macd_line = ema_fast = ema_slow
    signal_line = macd_line.ewm(span = signal, adjust = False).mean()
    hist = macd_line - signal_line
    return macd_line, signal_line, hist
````

````python
def add_llm_features(df, atr_col):

    out = df.copy().sort_values("time").reset_index(drop = True)

    # Price / trend indicators
    out["sma_20"] = out["close"].rolling(20).mean()
    out["sma_50"] = out["close"].rolling(50).mean()
    out["sma_200"] = out["close"].rolling(200).mean()

    out["ema_12"] = out["close"].ewm(span = 12, adjust = False).mean()
    out["ema_20"] = out["close"].ewm(span = 20, adjust = False).mean()
    out["ema_26"] = out["close"].ewm(span = 26, adjust = False).mean()
    out["ema_50"] = out["close"].ewm(span = 50, adjust = False).mean()
    out["ema_200"] = out["close"].ewm(span = 200, adjust = False).mean()

    # RSI / MACD
    out["rsi_14"] = compute_rsi(out["close"], window = 14)
    out["macd"], out["macd_signal"], out["macd_hist"] = compute_macd(out["close"])

    # ATR
    out["atr_pct"] = out[atr_col] / out["close"]

    # Return and momentum features
    out["ret_1d"] = out["close"].pct_change(1)
    out["ret_3d"] = out["close"].pct_change(3)
    out["ret_7d"] = out["close"].pct_change(7)
    out["ret_14d"] = out["close"].pct_change(14)

    out["mom_3d"] = out["close"] / out["close"].shift(3) - 1
    out["mom_7d"] = out["close"] / out["close"].shift(7) - 1
    out["mom_14d"] = out["close"] / out["close"].shift(14) - 1

    # Volatility / volume
    out["realized_vol_14"] = out["ret_1d"].rolling(14).std() * np.sqrt(14)
    out["volume_sma_20"] = out["volume"].rolling(20).mean()
    out["volume_zscore_20"] = ((out["volume"] - out["volume"].rolling(20).mean()) / out["volume"].rolling(20).std())

    # Distance to moving averages
    out["dist_to_sma20"] = (out["close"] / out["sma_20"]) - 1
    out["dist_to_sma50"] = (out["close"] / out["sma_50"]) - 1
    out["dist_to_ema20"] = (out["close"] / out["ema_20"]) - 1
    out["dist_to_ema50"] = (out["close"] / out["ema_50"]) - 1
    out["trend_spread_20_50"] = out["ema_20"] - out["ema_50"]
    out["trend_spread_50_200"] = out["ema_50"] - out["ema_200"]

    # Drawdown / breakout thresholds
    out["rolling_peak"] = out["close"].cummax()
    out["drawdown_from_peak"] = (out["rolling_peak"] - out["close"]) / out["rolling_peak"]
    out["rolling_low_30"] = out["close"].rolling(30).min()
    out["rolling_high_30"] = out["close"].rolling(30).max()
    out["breakout_pct_30"] = (out["close"] - out["rolling_high_30"]) / out["rolling_high_30"]
    out["rebound_from_low_30"] = (out["close"] - out["rolling_low_30"]) / out["rolling_low_30"]

    # Bitcoin-specific network feature transforms
    for col in ["hash_rate", "avg_block_size", "mempool_size"]:
        if col not in out.columns:
            out[col] = np.nan

        out[f"{col}_chg_7d"] = out[col].pct_change(7)
        out[f"{col}_zscore_30"] = ((out[col] - out[col].rolling(30).mean()) / out[col].rolling(30).std)
        
    # Composite custom features
    out["trend_score"] = (0.35 * (out["ema_20"] > out["ema_50"]).astype(float) + 0.35 * (out["ema_50"] > out["ema_200"]).astype(float)
    + 0.30 * (out["macd_hist"] > 0).astype(float))

    out["mean_reversion_score"] = ((out["rsi_14"] < 35).astype(float) + (out["drawdown_from_peak"] > 0.10).astype(float) +
    (out["dist_to_sma20"] < -0.03).astype(float)) / 3.0

    out["onchain_strength_score"] = ((out["hash_rate_chg_7d"] > 0).astype(float) + (out["avg_block_size_chg_7d"] > 0).astype(float)
    + (out["mempool_size_zscore_30"] > 0).astype(float)) / 3.0

    return out
````

````python
def prepare_market_context(days_back = 300):

    candles_df = get_btc_historical_candles(product_id = "BTC-USD", granularity = 86400, days_back = days_back)
    candles_df = calculate_atr(candles_df, window = 14)

    onchain_df = get_btc_onchain_features(days_back = days_back)

    merged = candles_df.merge(onchain_df, on = "time", how = "left")
    merged = merged.sort_values("time").reset_index(drop = True)

    # Forward fill slow-moving network metrics
    for col in ["hash_rate", "avg_block_size", "mempool_size"]:
        if col in merged.columns:
            merged[col] = merged[col].ffill()

    merged = add_llm_features(merged)
    merged["regime"] = merged.apply(detect_regime, axis = 1)

    return merged
````


## Hybrid LLM + Quantitative Trading System

This version uses the local LLaMA/Ollama model only for slower strategic decisions:

1. **Regime switching**: choose "value_investing" or "swing_trading" for the next review window.
2. **Feature selection**: identify which indicators/features matter most in the current market context.
3. **Risk adjustment**: adapt buy size, stop-loss width, take-profit distance, and whether new entries are allowed.

Per-candle execution remains deterministic and rule-based. This avoids slow, noisy LLM calls on every candle.


````python
# Call Ollama model and return parsed json

def query_ollama(prompt, model, url = "http://localhost:11434/api/generate", timeout = 180):
    
    payload = {"model": model, "prompt": prompt, "stream": False,
               "options": {"temperature": 0.1, "num_predict": 250}}

    try:
        resp = requests.post(url, json=payload, timeout=timeout)
        resp.raise_for_status()
        data = resp.json()
        raw = data.get("response", "").strip()
        if not raw:
            return None
        return json.loads(raw)
    except Exception as e:
        print(f"Warning: local LLM call failed; using heuristic fallback. Error: {e}")
        return None
````

````python
# Hybrid LLM + quantitative strategy

HYBRID_FEATURES = [
    "close", "ATR", "atr_pct", "rsi_14",
    "macd", "macd_signal", "macd_hist",
    "sma_20", "sma_50", "sma_200",
    "ema_20", "ema_50", "ema_200",
    "ret_3d", "ret_7d", "ret_14d",
    "mom_7d", "mom_14d",
    "realized_vol_14", "volume_zscore_20",
    "drawdown_from_peak", "dist_to_sma20", "dist_to_sma50",
    "trend_score", "mean_reversion_score",
    "hash_rate_chg_7d", "avg_block_size_chg_7d", "mempool_size_zscore_30",
    "onchain_strength_score",
]
````

````python
# Convert values returned by the LLM into floating point numbers

def safe_float(value, default):
    
    try:
        if value is None:
            return default
        value = float(value)
        if np.isnan(value) or np.isinf(value):
            return default
        return value
    except (TypeError, ValueError):
        return default
````

````python
# Parse boolean values returned by the LLM

def safe_bool(value, default = False):

    if isinstance(value, bool):
        return value
    if value is None:
        return default
    if isinstance(value, str):
        return value.strip().lower() in {"true", "yes", "1", "y"}
    return bool(value)
````

````python
# Clamp threshold values

def clamp(value, low, high):
    return max(low, min(high, value))
````

````python
# Create a summary of last N candles to feed to LLM

def summarize_recent_market_window(df, end_idx, lookback = 30, feature_cols = None):
    
    feature_cols = feature_cols or HYBRID_FEATURES
    start_idx = max(0, end_idx - lookback + 1)
    window = df.iloc[start_idx:end_idx + 1].copy()
    latest = window.iloc[-1]

    latest_snapshot = {}
    for col in feature_cols:
        if col in latest.index:
            val = latest[col]
            if pd.isna(val):
                latest_snapshot[col] = None
            elif isinstance(val, (np.integer, int)):
                latest_snapshot[col] = int(val)
            elif isinstance(val, (np.floating, float)):
                latest_snapshot[col] = round(float(val), 6)
            else:
                latest_snapshot[col] = str(val)

    def last_valid(col: str, default=None):
        if col not in window.columns:
            return default
        s = window[col].dropna()
        if len(s) == 0:
            return default
        return s.iloc[-1]

    summary = {
        "time": str(latest.get("time", "")),
        "lookback_days": int(len(window)),
        "latest": latest_snapshot,
        "window_stats": {
            "price_return": safe_float(window["close"].iloc[-1] / window["close"].iloc[0] - 1, 0.0) if len(window) > 1 else 0.0,
            "price_max_drawdown": safe_float(((window["close"].cummax() - window["close"]) / window["close"].cummax()).max(), 0.0),
            "avg_atr_pct": safe_float(window["atr_pct"].mean(), 0.0) if "atr_pct" in window.columns else 0.0,
            "avg_rsi_14": safe_float(window["rsi_14"].mean(), 50.0) if "rsi_14" in window.columns else 50.0,
            "latest_macd_hist": safe_float(last_valid("macd_hist", 0.0), 0.0),
            "latest_trend_score": safe_float(last_valid("trend_score", 0.0), 0.0),
            "latest_mean_reversion_score": safe_float(last_valid("mean_reversion_score", 0.0), 0.0),
            "latest_onchain_strength_score": safe_float(last_valid("onchain_strength_score", 0.0), 0.0),
        },
        "baseline_rule_regime": str(latest.get("regime", "value_investing")),
    }
    return summary
````

````python
# Build LLM prompt

def build_hybrid_llm_prompt(market_summary):
    schema = {"target_regime": "value_investing | swing_trading", "confidence": "float between 0 and 1",
              "selected_features": ["list", "of", "most", "useful", "features"], "reasoning": "brief explanation",
              "risk_adjustment": {"dca_buy_multiplier": "numeric, 0.5 to 1.5", "swing_buy_multiplier": "numeric, 0.5 to 1.5",
                                  "stop_atr_multiplier": "numeric, 1.25 to 3.5", "take_profit_multiplier": "numeric, 0.75 to 1.5",
                                  "max_position_multiplier": "numeric, 0.25 to 1.0", "allow_new_entries": "true or false",
                                  "force_reduce_exposure": "true or false"}}

    prompt = f"""You are a local open-source LLaMA crypto strategy supervisor.
    You do NOT execute trades directly. 
    Your job is only to set the strategy context for the next review window.
    The deterministic trading rules will execute every candle.
    Choose exactly one target_regime:
    - value_investing: BTC is weak, oversold, or in drawdown; prefer DCA-style accumulation only after rule-based pullback triggers.
    - swing_trading: BTC is trending upward; allow rule-based trend-following entries.

    Return ONLY valid JSON. Do not use markdown.
    All risk_adjustment fields must be numeric or boolean. Never return null.

    Market summary:
    {json.dumps(market_summary, indent=2)}

    Return schema:
    {json.dumps(schema, indent=2)}
    """.strip()
    return prompt
````
